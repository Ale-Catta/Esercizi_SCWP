# S3: A perfect circle

```python
import numpy as np
import matplotlib.pyplot as plt

# plt.rcParams.update({
#     "text.usetex": True,
#     "font.family": "Helvetica"
# })
```

```python
R = 10
theta = np.linspace(0,2*np.pi,1000)

x = R*np.cos(theta)
y = R*np.sin(theta)

fig = plt.figure()
ax = fig.add_subplot(111)
ax.plot(x, y, ".")
ax.set_aspect('equal')
```

```python
fig = plt.figure()
ax = fig.add_subplot(111)
ax.plot(x, y, ".r")
ax.set_aspect('equal')
ax.fill(x, y, color="blue", alpha=0.3)
```

```python
def draw_circle(x0, y0, R, color):
    theta = np.linspace(0,2*np.pi,1000)
    x = x0 + R*np.cos(theta)
    y = y0 + R*np.sin(theta)
    
    fig = plt.figure()
    ax = fig.add_subplot(111)
    ax.set_aspect('equal')
    ax.fill(x, y, color=color, alpha=0.3)
```

```python
draw_circle(1,1,10,"red")
```

```python
def draw_circle(x0, y0, R, color, fig=None, ax=None):

    theta = np.linspace(0, 2*np.pi, 1000)
    x = x0 + R*np.cos(theta)
    y = y0 + R*np.sin(theta)
    
    if ax is None:
        fig, ax = plt.subplots()

    ax.set_aspect('equal')
    ax.fill(x, y, color=color, alpha=0.3)
```

```python
x=np.linspace(0,2*np.pi,1000)
y=np.sin(x)

fig = plt.figure()
ax = fig.add_subplot(111)
ax.plot(x, y)
draw_circle(3,0,0.5,"red",fig,ax)
```

```python
import random

fig, ax = plt.subplots()

for _ in range(10):
    x0=random.uniform(-50, 50)  
    y0=random.uniform(-50, 50)  
    R=random.uniform(5, 20)
    color=[random.random() for _ in range(3)]
    
    draw_circle(x0, y0, R, color, fig=fig, ax=ax)

plt.show()
```

# S5: Subplots

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def derivative(x,f):
    delta_x = np.diff(x)
    delta_f = np.diff(f)
    derivative = delta_f / delta_x
    derivative = np.append(derivative, np.nan) #ultimo valore non importante ma necessario per avere array sempre della stessa lunghezza
    
    return derivative
```

```python
x = np.linspace(0,2*np.pi,100)

f = np.sin(x)
f_primo = derivative(x,f)
f_secondo = derivative(x,f_primo)
```

```python
fig = plt.figure()

ax1 = fig.add_subplot(311)

ax1.plot(x, f)
ax1.set_xlabel("x")
ax1.set_ylabel("f", fontsize="large")

ax2 = fig.add_subplot(312)

ax2.plot(x, f_primo)
ax2.set_xlabel("x")
ax2.set_ylabel("f'")

ax3 = fig.add_subplot(313)

ax3.plot(x, f_secondo)
ax3.set_xlabel("x")
ax3.set_ylabel("f''")

# tight_layout() makes sure things don't overlap
fig.tight_layout()
```
